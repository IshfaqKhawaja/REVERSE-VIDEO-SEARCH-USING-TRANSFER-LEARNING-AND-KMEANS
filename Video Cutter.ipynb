```
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import os
import cv2
from sklearn import tree
```

```
tree '/content/drive/MyDrive/New Data Set'
```

```
input_path = "/content/drive/MyDrive/New Data Set/"
output_path = "/content/drive/MyDrive/Abuzer/data/videos/"
```

```
def get_duration(video):
  # print(video)
  cap = cv2.VideoCapture(video)
  fps = cap.get(cv2.CAP_PROP_FPS)      # OpenCV2 version 2 used "CV_CAP_PROP_FPS"
  frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
  # print(fps)
  duration = frame_count/fps
  cap.release()
  return duration


print(get_duration("/content/drive/MyDrive/New Data Set/Cricket/c1"))
```

```
#Time gap is the gap or duration of video we want cut out
time_gap = 50

for category in os.listdir(input_path):
  print(category)
  out_p = os.path.join(output_path,category)
  input_p = os.path.join(input_path,category)
  if(category not in os.listdir(output_path)):
    os.mkdir(out_p)
  for video in os.listdir(input_p):
    # print(video)
    try:
      duration = int(get_duration(os.path.join(input_p,video)))
    except:
      print(video)
      continue
    # print(duration)
    start_time = duration//2
    end_time =  (duration//2) + time_gap if duration-time_gap>0 else duration
    try:
      ffmpeg_extract_subclip(os.path.join(input_p,video), start_time, end_time, targetname=os.path.join(out_p, video+".mp4"))
    except:
      print(video)
      continue



```
